```python
import pandas as pd
```

```python
steam_reviews = pd.read_csv('steam_reviews.csv')
```

```python
steam_reviews.head()
```

```python
steam_reviews = steam_reviews[['date_posted', 'funny', 'helpful', 'hour_played', 'recommendation', 'review', 'title']]
```

```python
steam_reviews.head()
```

```python
steam_reviews['id'] = steam_reviews.index
```

```python
steam_reviews.head()
```

```python
steam_reviews = steam_reviews.reindex(columns = ['id','title','date_posted','funny','helpful','hour_played','recommendation','review'])
```

```python
steam_reviews.isnull().any()
```

```python
steam_reviews['review'].fillna('Not shown', inplace = True)
```
