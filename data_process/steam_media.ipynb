```python
steam_screenshots = steam_media[['steam_appid', 'screenshots']]
steam_screenshots.head()
```

```python
final_dict = {'steam_appid':[0], 'thumbnail':['string'], 'full_image':['string']}
final_df = pd.DataFrame(final_dict)
final_df
```

```python
final_df = final_df.drop([0])
```

```python
final_df
```

```python
steam_screenshots.shape
```

```python
def get_http(a):
    thumbnail = []
    full = []
    while '{' in a:
        left = a.index('{')
        right = a.index('}')
        now = a[left + 1:right]
        sep1 = now.index('path_thumbnail')  + 18
        sep2 = now.index('path_full') + 13
        temp1 = now[sep1:sep2 - 17]
        temp2 = now[sep2:-1]
        thumbnail.append(temp1)
        full.append(temp2)
        a = a[right + 1:]
    return thumbnail, full
```

```python
for i in range(steam_screenshots.shape[0]):
    if i % 100 == 0:
        print(i)
    string = steam_screenshots['screenshots'][i]
    appid = steam_screenshots['steam_appid'][i]
    thumbnail, full = get_http(string)
    for j in range(len(full)):
        s = pd.Series([appid, thumbnail[j], full[j]], index = final_df.columns)
        final_df = final_df.append(s, ignore_index=True)
```
